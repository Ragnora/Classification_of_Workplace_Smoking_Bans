**IMPORT LIBRARY**

```python
import pandas as pd
import math
import plotly.express as px

from scipy import stats
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.naive_bayes import GaussianNB
```

**IMPORT DATASET**

```python
df = pd.read_csv('SmokeBan.csv')
df
```

**INPUT DAN TARGET**

```python
X = df.drop(['Unnamed: 0', 'smoker', 'hispanic'], axis=1)
y = df.smoker
print(X)
```

**CEK DATA NULL**

```python
X.columns[X.isna().any()]
```

**MENGUBAH STRING MENJADI NUMERIC**

```python
le = LabelEncoder()
X.ban = le.fit_transform(X.ban)
X.education = le.fit_transform(X.education)
X.afam = le.fit_transform(X.afam)
X.gender = le.fit_transform(X.gender)
print(X)
```

**SPLIT DATA TEST**

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=100)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

**FEATURE SCALING**

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
print(X_train)
```

**MODEL NAIVE BAYES**

```python
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

**PREDICT RESULT**

```python
y_pred = classifier.predict(X_test)
print(y_pred)
```

```python
print(y_test)
```

**AKURASI**

```python
cm = confusion_matrix(y_test, y_pred)
ac = accuracy_score(y_test, y_pred)
print("Accuracy:", (ac * 100), "%")
```

MEAN

```python
def mean(dataset, kolom):
    df = dataset[kolom]
    n = len(df)
    mean = sum(df)/n
    return mean

#Laki-Laki
filtered_dataset = df[(df['smoker'] == 'yes') & (df['gender'] == 'male')]
mean_age = mean(filtered_dataset, 'age')
print("Rata-Rata Usia Perokok Laki-Laki: ", mean_age)

#Perempuan
filtered_dataset = df[(df['smoker'] == 'yes') & (df['gender'] == 'female')]
mean_age = mean(filtered_dataset, 'age')
print("Rata-Rata Usia Perokok Perempuan: ", mean_age)

#Keseluruhan
filtered_dataset = df[(df['smoker'] == 'yes')]
mean_age = mean(filtered_dataset, 'age')
print("Rata-Rata Usia Perokok Keseluruhan: ", mean_age)
```

MODUS

```python
def modus(dataset, kolom):
    df = dataset[kolom]
    modus = df.mode().values[0]
    return modus

#Laki-Laki
filtered_dataset = df[(df['smoker'] == 'yes') & (df['gender'] == 'male')]
modus_age = modus(filtered_dataset, 'age')
modus_edu = modus(filtered_dataset, 'education')
modus_afm = modus(filtered_dataset, 'afam')
print("Modus Usia Perokok Laki-Laki: ", modus_age)
print("Modus Edukasi Perokok Laki-Laki: ", modus_edu)
print("Modus Afam Perokok Laki-Laki: ", modus_afm)

#Perempuan
filtered_dataset = df[(df['smoker'] == 'yes') & (df['gender'] == 'female')]
modus_age = modus(filtered_dataset, 'age')
modus_edu = modus(filtered_dataset, 'education')
modus_afm = modus(filtered_dataset, 'afam')
print("\nModus Usia Perokok Perempuan: ", modus_age)
print("Modus Edukasi Perokok Perempuan: ", modus_edu)
print("Modus Afam Perokok Perempuan: ", modus_afm)

#Keseluruhan
filtered_dataset = df[(df['smoker'] == 'yes')]
modus_age = modus(filtered_dataset, 'age')
modus_edu = modus(filtered_dataset, 'education')
modus_afm = modus(filtered_dataset, 'afam')
print("\nModus Usia Perokok Keseluruhan: ", modus_age)
print("Modus Edukasi Perokok Keseluruhan: ", modus_edu)
print("Modus Afam Perokok Keseluruhan: ", modus_afm)
```

MEDIAN

```python
def median(dataset, kolom):
    df = dataset[kolom]
    sorted_data = sorted(df)
    n = len(sorted_data)
    if n % 2 == 0:
        median = (sorted_data[n//2]+sorted_data[n//2-1])/2
    else:
        median = sorted_data[n//2]
    return median

#Laki-Laki
filtered_dataset = df[(df['smoker'] == 'yes') & (df['gender'] == 'male')]
median_age = median(filtered_dataset, 'age')
print("Median Usia Perokok Laki-Laki: ", median_age)

#Perempuan
filtered_dataset = df[(df['smoker'] == 'yes') & (df['gender'] == 'female')]
median_age = median(filtered_dataset, 'age')
print("Median Usia Perokok Perempuan: ", median_age)

#Keseluruhan
filtered_dataset = df[(df['smoker'] == 'yes')]
median_age = median(filtered_dataset, 'age')
print("Median Usia Perokok Keseluruhan: ", median_age)
```

QUARTILE

```python
def quartile(dataset, kolom):
    data = sorted(dataset[kolom])
    n = len(data)
    q1_index = int(n*0.25)
    q2_index = int(n*0.50)
    q3_index = int(n*0.75)
    quartil = [
        data[q1_index],
        data[q2_index],
        data[q3_index]
    ]
    return quartil

#Laki-Laki
filtered_dataset = df[(df['smoker'] == 'yes') & (df['gender'] == 'male')]
quartile_age = quartile(filtered_dataset, 'age')
# quartile_edu = quartile(filtered_dataset, 'education')
# quartile_afm = quartile(filtered_dataset, 'afam')
print("Quartile Usia Perokok Laki-Laki: ", quartile_age)
# print("Quartile Edukasi Perokok Laki-Laki: ", quartile_edu)
# print("Quartile Afam Perokok Laki-Laki: ", quartile_afm)

#Perempuan
filtered_dataset = df[(df['smoker'] == 'yes') & (df['gender'] == 'female')]
quartile_age = quartile(filtered_dataset, 'age')
# quartile_edu = quartile(filtered_dataset, 'education')
# quartile_afm = quartile(filtered_dataset, 'afam')
print("Quartile Usia Perokok Perempuan: ", quartile_age)
# print("Quartile Edukasi Perokok Perempuan: ", quartile_edu)
# print("Quartile Afam Perokok Perempuan: ", quartile_afm)

#Keseluruhan
filtered_dataset = df[(df['smoker'] == 'yes')]
quartile_age = quartile(filtered_dataset, 'age')
print("Quartile Usia Perokok Keseluruhan: ", quartile_age)
```

STANDAR DEVIASI

```python
def std(dataset, kolom):
    data = dataset[kolom]
    n = len(data)
    mean = sum(data)/n
    squared_diff_sum = sum((x-mean)**2 for x in data)
    std_dev = math.sqrt(squared_diff_sum/n)
    return std_dev

#Laki-Laki
filtered_dataset = df[(df['smoker'] == 'yes') & (df['gender'] == 'male')]
std_age = std(filtered_dataset, 'age')
print("Standar Deviasi Usia Perokok Laki-Laki: ", std_age)

#Perempuan
filtered_dataset = df[(df['smoker'] == 'yes') & (df['gender'] == 'female')]
std_age = std(filtered_dataset, 'age')
print("Standar Deviasi Usia Perokok Perempuan: ", std_age)

#Keseluruhan
filtered_dataset = df[(df['smoker'] == 'yes')]
std_age = std(filtered_dataset, 'age')
print("Standar Deviasi Usia Perokok Keseluruhan: ", std_age)
```

VARIANCE

```python
def var(dataset, kolom):
    data = dataset[kolom]
    n = len(data)
    mean = sum(data)/n
    squared_diff_sum = sum((x-mean)**2 for x in data)
    variance = squared_diff_sum/n
    return variance

#Laki-Laki
filtered_dataset = df[(df['smoker'] == 'yes') & (df['gender'] == 'male')]
var_age = var(filtered_dataset, 'age')
print("Variance Usia Perokok Laki-Laki: ", var_age)

#Perempuan
filtered_dataset = df[(df['smoker'] == 'yes') & (df['gender'] == 'female')]
var_age = var(filtered_dataset, 'age')
print("Variance Usia Perokok Perempuan: ", var_age)

#Keseluruhan
filtered_dataset = df[(df['smoker'] == 'yes')]
var_age = var(filtered_dataset, 'age')
print("Variance Usia Perokok Keseluruhan: ", var_age)
```

**NAIVE BAYES CLASSIFIER**

```python
df = pd.read_csv('SmokeBan.csv')

#Input dan Target
X = df.drop(['Unnamed: 0', 'smoker', 'hispanic'], axis=1)
y = df.smoker

#Mengubah Nilai String Menjadi Angka (Preprocessing)
le = LabelEncoder()
X.ban = le.fit_transform(X.ban)
X.education = le.fit_transform(X.education)
X.afam = le.fit_transform(X.afam)
X.gender = le.fit_transform(X.gender)

#SPLIT DATA TEST
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=100)
X_train.shape, X_test.shape, y_train.shape, y_test.shape

#NORMALISASI
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)

#MODEL NAIVE BAYES
classifier = GaussianNB()
classifier.fit(X_train, y_train)

#MENAMPUNG NILAI PREDIKSI
y_pred = classifier.predict(X_test)

#AKURASI DATA
cm = confusion_matrix(y_test, y_pred)
ac = accuracy_score(y_test, y_pred)
print("Accuracy:", (ac * 100), "%")
```
